```python
x=df.iloc[:,0:-1]
```

```python
x
```

```python
y=df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train.shape
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc=StandardScaler()
```

```python
x_train_scaled=sc.fit_transform(x_train)
```

```python
x_test_scaled=sc.transform(x_test)
```

```python
x_train_scaled
```

```python
from sklearn.linear_model import SGDClassifier
```

```python
sgd=SGDClassifier()
```

```python
sgd.fit(x_train_scaled,y_train)
```

```python
y_pred= sgd.predict(x_test_scaled)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,y_pred)
```

**gradient boost got best accuracy**

```python
from xgboost import XGBClassifier
```

```python
xgb=XGBClassifier()
```

```python
xgb.fit(x_train_scaled,y_train)
```

```python
y_pred=xgb.predict(x_test_scaled)
```

```python
accuracy_score(y_test,y_pred)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
```

```python
lr.fit(x_train_scaled,y_train)
```

```python
y_pred=lr.predict(x_test_scaled)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test,y_pred)
```

**Linear regression got less accuracy**

```python
from sklearn.ensemble import RandomForestClassifier
list1=[]
for i in range(1,101):
    classifier= RandomForestClassifier(n_estimators= i, criterion="entropy")  
    classifier.fit(x_train_scaled, y_train) 
    y_pred=classifier.predict(x_test_scaled)
    list1.append(accuracy_score(y_test,y_pred))
    

    
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.plot(range(1,101),list1)
```

**by this graph we get highest accuracy at 63**

# Again calculating for best accuracy
